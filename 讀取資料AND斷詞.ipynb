```python
#讀檔，初步分割資料
with open('/content/drive/MyDrive/佩玲論文資料/alldata_2021.txt','r') as fh:
  tmp = fh.read()
  onelinelist = tmp.replace("\n","")
  onelinelist = onelinelist.replace("/", '').replace(',', '').replace('?', '').replace('(', '').replace(')', '').replace('!', '').replace('"', '').replace('●', '').replace(':', '').replace('’', '').replace('•', '').replace('”', '').strip()
 #onelinelist = onelinelist.split("(?<![\d])[.]")
  onelinelist = re.split(r"(?<![\d])[.]", onelinelist)
  print(onelinelist)
```

```python
#初步看資料有幾筆
len(onelinelist)
```

```python
#去除開頭空白
for i in range(len(onelinelist)):
  onelinelist[i] = onelinelist[i].lstrip()
```
